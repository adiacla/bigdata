```python
#Estructura básica de Matplotlib.pyplot
fig, ax = plt.subplots(figsize=(4,3))
ax.hist(x = dfc.calefaccion, color="aqua")
plt.title('Histograma', loc = "left")
plt.title('Calefaccion', loc = "right")
ax.grid()
ax.set_xlabel("Instalacias de Calefaccion")
ax.set_ylabel("Cantidad en el dataset")
```

```python
#Grafico de histograma de todas las variables (object, string, category)

columnas_categoricas=dfc.select_dtypes(include='object').columns

#Definir los parámetros de la grafica
fig,axes=plt.subplots(nrows=3, ncols=3,figsize=(9,5))
axes=axes.flat


#recorrer todas las columnas
for i,columna in enumerate(columnas_categoricas):
  #print(pd.value_counts(dfc[columna], sort=True, ascending=True))
  dfc[columna].value_counts().plot.barh(ax=axes[i])
  axes[i].set_title(columna, fontsize=7,fontweight="bold")
  axes[i].tick_params(labelsize=6)
  axes[i].set_xlabel("")

#delaxes borrar los subplots que se deseen
fig.delaxes(axes[7])
fig.delaxes(axes[8])

#Ajustar los tamaños automáticamente para que las gráficas no queden montados
fig.tight_layout()
fig.subplots_adjust(top=0.8)
fig.suptitle("Distribución de las variables cualitativas", fontsize=10,fontweight="bold")

```
